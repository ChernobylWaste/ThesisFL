## 4. Data Preprocessing

```python
# Freeing Memory Computer

old_memory_usage = sampled_data.memory_usage().sum() / 1024 ** 2
print(f'Initial memory usage: {old_memory_usage:.2f} MB')
for col in sampled_data.columns:
    col_type = sampled_data[col].dtype
    if col_type != object:
        c_min = sampled_data[col].min()
        c_max = sampled_data[col].max()
        # Downcasting float64 to float32
        if str(col_type).find('float') >= 0 and c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
            sampled_data[col] = sampled_data[col].astype(np.float32)

        # Downcasting int64 to int32
        elif str(col_type).find('int') >= 0 and c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
            sampled_data[col] = sampled_data[col].astype(np.int32)

new_memory_usage = sampled_data.memory_usage().sum() / 1024 ** 2
print(f"Final memory usage: {new_memory_usage:.2f} MB")

print(f'Reduced memory usage: {1 - (new_memory_usage / old_memory_usage):.2%}')
```

```python
# 4.5. Scaling

# Sampled data masih memiliki attack type, akan didrop dan dijadikan variabel attacks. Setelah itu tidak ada lagi fitur yang berbentuk object, maka akan dilakukan scaling menggunakan minmaxscaler
# MinMaxScaler melakukan scaling dengan membagi semua nilai dengan nilai tertinggi pada fitur, mengubah seluruh angka menjadi antara 0 sampai 1

features = sampled_data.drop('Attack Type', axis = 1)
attacks = sampled_data['Attack Type']

scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(features)
```

```python
# Train Data Creation

# Membuat variable baru dinamakan new_data yang akan dipakai untuk training, data ini telah discaling dan akan mengembalikan Attack Type menjadi fitur pada data ini

new_data = pd.DataFrame(scaled_features)
new_data['Attack Type'] = attacks.values
```

```python
new_data
```

```python
# Dimension Reduction

# Dengan menggunakan PCA dapat mengurangi dimensi data untuk mempermudah mesin mempelajarinya
# Pengurangan dimensi ini dilakukan dengan kehilangan data seminimal mungkin

size = len(features.columns) // 2
ipca = IncrementalPCA(n_components = size, batch_size = 500)
for batch in np.array_split(scaled_features, len(features) // 500):
    ipca.partial_fit(batch)

transformed_features = ipca.transform(scaled_features)
new_data = pd.DataFrame(transformed_features, columns = [f'PC{i+1}' for i in range(size)])
new_data['Attack Type'] = attacks.values

print(f'information retained: {sum(ipca.explained_variance_ratio_):.2%}')
```
